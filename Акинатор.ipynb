```python
custom_columns = {
    'Universe': ['DC', 'Marvel'],
    'Superpowers': ['Super Strength', 'Flight', 'Super Speed', 'Super Intelligence', 'Super Durability'],
    'Affiliation': ['Independent', 'Avengers', 'Guardians of the Galaxy', 'Injustice League', 'Justice League', 'Outlaw', 'Suicide Squad', 'Teen Titans', 'X-Force', 'X-Men'],
    'Gender': ['Female', 'Male'],
    'Alignment': ['Evil', 'Good', 'Neutral'],
    'Species': ['Amazonian', 'Asgardian', 'Atlantean', 'Flora Colossus', 'Human', 'Hybrid', 'Kryptonian', 'Mutant', 'Zehoberei'],
    'Occupation': ['Assassin', 'Billionaire', 'Burglar', 'Child', 'Criminal', 'Electrical Engineer', 'God of Thunder', 'Guardian of the Galaxy', 'Journalist', 'King', 'Photographer', 'Pilot', 'Princess', 'Psychiatrist', 'Reporter', 'Scientist', 'Soldier', 'Space Explorer', 'Student', 'Surgeon', 'Teacher', 'Trickster God', 'Vigilante', 'Witch'],
    'Location': ['Asgard', 'Atlantis', 'Brooklyn', 'Canada', 'Central City', 'Coast City', 'Dayton', 'Earth', 'Fawcett City', 'Gotham City', 'Kenya', 'Moscow', 'Neopolis', 'New York City', 'Planet X', 'Smallville', 'Star City', 'Themyscira', 'Transia', 'Wakanda', 'Zehoberei.1'],
}

questions = {
    'Universe': 'Is your character from the universe',
    'Superpowers': 'Does your character possess any superpowers like',
    'Affiliation': 'Is your character affiliated with team, such as',
    'Gender': 'Is your character',
    'Alignment': 'Would you consider your character to be',
    'Species': 'The species of your character is',
    'Occupation': 'The occupation of your character is',
    'Location': 'Is your character was born in'
}

def get_user_response():
    user_response = input().lower().strip()
    valid_responses = ['yes', 'probably yes', 'no', 'probably no', "i don't know", 'quit']

    if user_response in valid_responses:
        return user_response
    else:
        print("Invalid response. Please enter 'yes', 'no', 'probably yes', 'probably no', 'I don't know', or 'quit'.")
        return get_user_response()

def play_akinator_tree(model, label_encoder):
    while True:
        current_node = 0  # Start at the root node

        while model.tree_.children_left[current_node] != model.tree_.children_right[current_node]:
            # Ask a question based on the current node
            feature = X.columns[model.tree_.feature[current_node]]

            for question_group, related_features in custom_columns.items():
                if feature in related_features:
                    user_response = input(f"{questions[question_group]} {feature.capitalize()}? (Yes/No/I don't know/Probably yes/Probably no/Quit): ")
                    break

            if user_response.lower() == 'quit':
                print("Quitting the game. Goodbye!")
                return

            if user_response.lower() == 'yes' or user_response == '1':
                    current_node = clf.tree_.children_right[current_node]
            elif user_response.lower() == 'probably yes' or user_response == '4':
                    current_node = clf.tree_.children_right[current_node]
            elif user_response.lower() == 'no' or user_response == '2':
                    current_node = clf.tree_.children_left[current_node]
            elif user_response.lower() == 'probably no' or user_response == '5':
                    current_node = clf.tree_.children_left[current_node]
            elif user_response.lower() == "i don't know" or user_response == '3':
                    current_node += 1
            elif user_response.lower() == 'quit' or user_response == '6':
                    print("Quitting the game. Goodbye!")
                    

        # Leaf node reached, make a prediction
        predicted_class = model.tree_.value[current_node].argmax()
        predicted_name = label_encoder.inverse_transform([predicted_class])[0]
        print(f"My guess is {predicted_name}. Was I correct? (yes/no)")

        user_response = input().lower()
        if user_response == 'yes':
            print("I guessed it right!")
        else:
            print("Oops! ooks like you defeated me !.")

        # Ask if the player wants to play again
        play_again = input("Do you want to play again? (yes/no): ").lower()
        if play_again != 'yes':
            print("Thanks for playing! Goodbye!")
            break
```

```python
def print_decision_tree_structure(tree, feature_names):
    n_nodes = tree.tree_.node_count
    children_left = tree.tree_.children_left
    children_right = tree.tree_.children_right
    feature = tree.tree_.feature
    threshold = tree.tree_.threshold

    for i in range(n_nodes):
        if children_left[i] != children_right[i]:  # non-leaf node
            feature_name = feature_names[feature[i]]
            print(f"Node {i}: Ask about {feature_name} (threshold: {threshold[i]})")
        else:  # leaf node
            value = tree.tree_.value[i].argmax()
            print(f"Node {i}: Predict {value}")

# Call this function with your decision tree model and feature names
print_decision_tree_structure(clf, X.columns)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

class AkinatorTree:
    def __init__(self, dataset=None):
        self.df = pd.read_csv(dataset)
        self.custom_columns = {
            'Universe': ['DC', 'Marvel'],
            'Superpowers': ['Super Strength', 'Flight', 'Super Speed', 'Super Intelligence', 'Super Durability'],
            'Affiliation': ['Independent', 'Avengers', 'Guardians of the Galaxy', 'Injustice League', 'Justice League', 'Outlaw', 'Suicide Squad', 'Teen Titans', 'X-Force', 'X-Men'],
            'Gender': ['Female', 'Male'],
            'Alignment': ['Evil', 'Good', 'Neutral'],
            'Species': ['Amazonian', 'Asgardian', 'Atlantean', 'Flora Colossus', 'Human', 'Hybrid', 'Kryptonian', 'Mutant', 'Zehoberei'],
            'Occupation': ['Assassin', 'Billionaire', 'Burglar', 'Child', 'Criminal', 'Electrical Engineer', 'God of Thunder', 'Guardian of the Galaxy', 'Journalist', 'King', 'Photographer', 'Pilot', 'Princess', 'Psychiatrist', 'Reporter', 'Scientist', 'Soldier', 'Space Explorer', 'Student', 'Surgeon', 'Teacher', 'Trickster God', 'Vigilante', 'Witch'],
            'Location': ['Asgard', 'Atlantis', 'Brooklyn', 'Canada', 'Central City', 'Coast City', 'Dayton', 'Earth', 'Fawcett City', 'Gotham City', 'Kenya', 'Moscow', 'Neopolis', 'New York City', 'Planet X', 'Smallville', 'Star City', 'Themyscira', 'Transia', 'Wakanda', 'Zehoberei.1'],
        }

        self.questions = {
            'Universe': 'Is your character from the universe',
            'Superpowers': 'Does your character possess any superpowers like',
            'Affiliation': 'Is your character affiliated with a team, such as',
            'Gender': 'Is your character',
            'Alignment': 'Would you consider your character to be',
            'Species': 'The species of your character is',
            'Occupation': 'The occupation of your character is',
            'Location': 'Is your character born in'
        }

        self.clf = DecisionTreeClassifier()
        self.label_encoder = LabelEncoder()
        self.X, self.y = self.create_features_and_target()

    def create_features_and_target(self):
        X = self.df.drop(['name'], axis=1)
        y = self.df['name']
        return X, y

    def train_model(self):
        self.label_encoder.fit(self.y)  
        self.clf.fit(self.X, self.label_encoder.transform(self.y))  

    def get_user_response(self):
        user_response = input().lower().strip()
        valid_responses = ['yes', 'probably yes', 'no', 'probably no', "i don't know", 'quit',]

        if user_response in valid_responses:
            return user_response
        else:
            # print("Invalid response. Please enter 'yes', 'no', 'probably yes', 'probably no', 'I don't know', or 'quit'.")
            return self.get_user_response()

    def play_akinator_tree(self):
        while True:
            current_node = 0

            while self.clf.tree_.children_left[current_node] != self.clf.tree_.children_right[current_node]:
                feature = self.X.columns[self.clf.tree_.feature[current_node]]

                for question_group, related_features in self.custom_columns.items():
                    if feature in related_features:
                        # user_response = input(f"{self.questions[question_group]} {feature.capitalize()}? (Yes/No/I don't know/Quit): ")
                        user_response = input(f"{self.questions[question_group]} {feature.capitalize()}? (Yes/No/I don't know/Probably yes/Probably no/Quit): ")
                        break
                else:
                    user_response = input(f"{feature.capitalize()}? (Yes/No/I don't know/Quit): ")

                if user_response.lower() == 'quit':
                    print("Quitting the game. Goodbye!")
                    return

                if user_response.lower() == 'yes' or user_response == '1':
                    current_node = self.clf.tree_.children_right[current_node]
                elif user_response.lower() == 'probably yes' or user_response == '4':
                    current_node = self.clf.tree_.children_right[current_node]
                elif user_response.lower() == 'no' or user_response == '2':
                    current_node = self.clf.tree_.children_left[current_node]
                elif user_response.lower() == 'probably no' or user_response == '5':
                    current_node = self.clf.tree_.children_left[current_node]
                elif user_response.lower() == "i don't know" or user_response == '3':
                    current_node += 1
                elif user_response.lower() == 'quit' or user_response == '6':
                    print("Quitting the game. Goodbye!")
                    return

            predicted_class = self.clf.tree_.value[current_node].argmax()
            predicted_name = self.label_encoder.inverse_transform([predicted_class])[0]
            print(f"My guess is {predicted_name}. Was I correct? (yes/no)")

            user_response = input().lower()No
            if user_response == 'yes':
                print("I guessed it right!")
            else:
                print("Oops! Looks like you defeated me!")

            play_again = input("Do you want to play again? (yes/no): ").lower()
            if play_again != 'yes':
                print("Thanks for playing! Goodbye!")
                break
```
